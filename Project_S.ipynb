## Data cleaning

### Subtask:
Clean the data by handling missing values, converting data types, and addressing inconsistencies in the 'tahun' column.


**Reasoning**:
Based on the exploration, several columns have a large number of missing values. These columns are primarily related to 'penyusutan'. Since they have many missing values and might not be central to a general financial analysis, it's reasonable to drop them as per the instructions.



```python
columns_to_drop = [
    'penyusutan_total',
    'penyusutan_garis_lurus',
    'penyusutan_saldo_menurun',
    'penyusutan_fiskal',
    'laba_kotor_operasional' # This also has a large number of missing values
]

df = df.drop(columns=columns_to_drop)
display(df.isnull().sum())
```

**Reasoning**:
Some columns still have missing values, but fewer than the ones just dropped. Imputation is a suitable approach for these as dropping rows might remove too much data. Since these are financial figures, the median is a more robust imputation strategy than the mean, as it is less affected by outliers. Columns with remaining large number of NaNs like `overhead_farmasi` and subsequent are also financial and related, hence they will be imputed with median as well.



```python
columns_to_impute_median = [
    'tahun', # Will be handled later, but include for now
    'persediaan_awal',
    'persediaan_akhir',
    'hpp_rawat_inap',
    'hpp_rawat_jalan',
    'total_hpp',
    'pendapatan_operasional',
    'pendapatan_penjualan',
    'ppn_keluaran',
    'ppn_masukan',
    'ppn_netto',
    'laba_kotor_penjualan',
    'beban_operasional',
    'overhead_farmasi',
    'laba_sebelum_pajak',
    'laba_kena_pajak',
    'tarif_pph_normal',
    'pph_badan_normal',
    'tarif_pph_tax_holiday',
    'pph_badan_tax_holiday',
    'laba_bersih_normal',
    'laba_bersih_tax_holiday',
    'kerugian_fiskal',
    'tax_holiday_status' # Will be handled later, but include for now
]

for col in columns_to_impute_median:
    # Attempt to convert to numeric first, coercing errors
    df[col] = pd.to_numeric(df[col], errors='coerce')
    # Then impute the median
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)

display(df.isnull().sum())
```

**Reasoning**:
The previous step attempted to convert all specified columns to numeric and then impute, but it seems like some columns still have missing values after imputation. This indicates that the `pd.to_numeric` conversion with `errors='coerce'` might have turned non-numeric values into NaN, and the subsequent median imputation didn't fill all of them, or there were other issues. It is necessary to handle the non-numeric characters in the columns that should be numeric before attempting conversion and imputation again. Also, the 'tahun' column and categorical columns like 'tarif_pph_normal' and 'tax_holiday_status' were included in the imputation list but need specific handling. The 'tahun' column needs non-year values removed/corrected, and 'tarif_pph_normal' and 'tax_holiday_status' are not suitable for median imputation after numerical conversion. Let's specifically target the columns that failed to impute properly and clean them by removing non-numeric characters before conversion and imputation. We will also address the 'tahun' column cleaning and handle 'tarif_pph_normal' and 'tax_holiday_status' separately.



```python
import numpy as np

# List of columns that should be numeric but contain non-numeric characters
numeric_cols_with_chars = [
    'persediaan_awal',
    'persediaan_akhir',
    'hpp_rawat_inap',
    'hpp_rawat_jalan',
    'total_hpp',
    'pendapatan_operasional',
    'pendapatan_penjualan',
    'ppn_keluaran',
    'ppn_masukan',
    'ppn_netto',
    'laba_kotor_penjualan',
    'beban_operasional',
    'overhead_farmasi',
    'laba_sebelum_pajak',
    'laba_kena_pajak',
    'pph_badan_normal',
    'tarif_pph_tax_holiday',
    'pph_badan_tax_holiday',
    'laba_bersih_normal',
    'laba_bersih_tax_holiday',
    'kerugian_fiskal'
]

for col in numeric_cols_with_chars:
    if col in df.columns:
        # Convert to string to handle potential mixed types, then remove '.' and ','
        df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
        # Convert to numeric, coercing errors to NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Impute missing values in these now-numeric columns with the median
for col in numeric_cols_with_chars:
     if col in df.columns and df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)


# Handle the 'tahun' column: remove non-year values and convert to integer
# Replace non-numeric values (like '22%', 'Normal') with NaN first
df['tahun'] = pd.to_numeric(df['tahun'], errors='coerce')
# Impute missing or non-numeric years with the mode (most frequent year)
if df['tahun'].isnull().sum() > 0:
    mode_tahun = df['tahun'].mode()[0] # mode() can return multiple values if tied, take the first
    df['tahun'] = df['tahun'].fillna(mode_tahun)
# Convert to integer type
df['tahun'] = df['tahun'].astype(int)

# Handle 'tarif_pph_normal' - it seems to be a constant '22%'. Convert to numeric 0.22 and impute if needed.
if 'tarif_pph_normal' in df.columns:
    df['tarif_pph_normal'] = df['tarif_pph_normal'].astype(str).str.replace('%', '', regex=False)
    df['tarif_pph_normal'] = pd.to_numeric(df['tarif_pph_normal'], errors='coerce') / 100
    if df['tarif_pph_normal'].isnull().sum() > 0:
        # Impute with the calculated value (0.22) or median if 0.22 is NaN
        median_tarif = df['tarif_pph_normal'].median()
        df['tarif_pph_normal'] = df['tarif_pph_normal'].fillna(median_tarif if pd.notna(median_tarif) else 0.22)


# Handle 'tax_holiday_status' - seems categorical, impute with mode
if 'tax_holiday_status' in df.columns and df['tax_holiday_status'].isnull().sum() > 0:
    mode_status = df['tax_holiday_status'].mode()[0]
    df['tax_holiday_status'] = df['tax_holiday_status'].fillna(mode_status)


display(df.isnull().sum())
display(df.head())
display(df.info())
```

**Reasoning**:
The traceback indicates a `KeyError: 0` when trying to access the mode of the 'tax_holiday_status' column using `.mode()[0]`. This likely means that the `.mode()` call returned an empty Series (no mode found), and attempting to access the element at index 0 failed. This could happen if the column is entirely NaN or has no clear mode after previous processing. We need to handle this case by checking if the mode is available before attempting to access it. Let's fix this specific error and re-run the cleaning process. We also need to check if the conversion and imputation for other columns worked correctly.



```python
import numpy as np

# List of columns that should be numeric but contain non-numeric characters
numeric_cols_with_chars = [
    'persediaan_awal',
    'persediaan_akhir',
    'hpp_rawat_inap',
    'hpp_rawat_jalan',
    'total_hpp',
    'pendapatan_operasional',
    'pendapatan_penjualan',
    'ppn_keluaran',
    'ppn_masukan',
    'ppn_netto',
    'laba_kotor_penjualan',
    'beban_operasional',
    'overhead_farmasi',
    'laba_sebelum_pajak',
    'laba_kena_pajak',
    'pph_badan_normal',
    'tarif_pph_tax_holiday',
    'pph_badan_tax_holiday',
    'laba_bersih_normal',
    'laba_bersih_tax_holiday',
    'kerugian_fiskal'
]

for col in numeric_cols_with_chars:
    if col in df.columns:
        # Convert to string to handle potential mixed types, then remove '.' and ','
        df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
        # Convert to numeric, coercing errors to NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Impute missing values in these now-numeric columns with the median
for col in numeric_cols_with_chars:
     if col in df.columns and df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val if pd.notna(median_val) else 0) # Impute with 0 if median is NaN

# Handle the 'tahun' column: remove non-year values and convert to integer
# Replace non-numeric values (like '22%', 'Normal') with NaN first
df['tahun'] = pd.to_numeric(df['tahun'], errors='coerce')
# Impute missing or non-numeric years with the mode (most frequent year)
if df['tahun'].isnull().sum() > 0:
    mode_tahun_series = df['tahun'].mode()
    if not mode_tahun_series.empty:
        mode_tahun = mode_tahun_series[0] # mode() can return multiple values if tied, take the first
        df['tahun'] = df['tahun'].fillna(mode_tahun)
    else:
         # If mode is empty (e.g., all NaNs), impute with a default like 2023
        df['tahun'] = df['tahun'].fillna(2023) # Assuming 2023 is a reasonable default

# Convert to integer type, after ensuring no NaNs
df['tahun'] = df['tahun'].astype(int)


# Handle 'tarif_pph_normal' - it seems to be a constant '22%'. Convert to numeric 0.22 and impute if needed.
if 'tarif_pph_normal' in df.columns:
    df['tarif_pph_normal'] = df['tarif_pph_normal'].astype(str).str.replace('%', '', regex=False)
    df['tarif_pph_normal'] = pd.to_numeric(df['tarif_pph_normal'], errors='coerce') / 100
    if df['tarif_pph_normal'].isnull().sum() > 0:
        # Impute with the calculated value (0.22) or median if 0.22 is NaN
        median_tarif = df['tarif_pph_normal'].median()
        df['tarif_pph_normal'] = df['tarif_pph_normal'].fillna(median_tarif if pd.notna(median_tarif) else 0.22)


# Handle 'tax_holiday_status' - seems categorical, impute with mode, handle case where mode is empty
if 'tax_holiday_status' in df.columns and df['tax_holiday_status'].isnull().sum() > 0:
    mode_status_series = df['tax_holiday_status'].mode()
    if not mode_status_series.empty:
        mode_status = mode_status_series[0]
        df['tax_holiday_status'] = df['tax_holiday_status'].fillna(mode_status)
    else:
        # If mode is empty (e.g., all NaNs), impute with a default like 'Unknown' or drop (decided to impute with 'Unknown')
        df['tax_holiday_status'] = df['tax_holiday_status'].fillna('Unknown')


display(df.isnull().sum())
display(df.head())
display(df.info())
```

Create colored line plots to visualize the trends of 'pendapatan_operasional', 'total_hpp', 'laba_sebelum_pajak', and 'laba_bersih_normal' over the years.


**Reasoning**:
Melt the `trends_by_year` DataFrame to a long format suitable for plotting with Seaborn and then create the line plot to visualize the trends of the specified financial metrics over the years.
