## Data cleaning

### Subtask:
Handle missing values in the DataFrame based on the information gathered during data exploration.


**Reasoning**:
Identify columns with a high percentage of missing values and drop columns with a very high number of missing values based on the previous exploration.



```python
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

high_missing_columns = missing_percentage[missing_percentage > 50].index.tolist()

print("Columns with percentage of missing values:")
print(missing_percentage)
print("\nColumns to be dropped due to high missing values:")
print(high_missing_columns)

df_cleaned = df.drop(columns=high_missing_columns)

display(df_cleaned.head())
```

**Reasoning**:
Handle missing values in the remaining columns. Based on the output of `df.info()` from the exploration phase, the remaining columns are primarily numerical and contain missing values. Impute the missing numerical values with the mean.



```python
for col in df_cleaned.columns:
    if df_cleaned[col].dtype in ['float64', 'int64']:
        df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mean())

print("\nMissing values after imputation:")
print(df_cleaned.isnull().sum())

display(df_cleaned.head())
```

**Reasoning**:
The previous imputation did not fully handle the missing values, likely due to the data type being 'object' instead of a numerical type, despite the columns containing numbers with thousands separators ('.'). Convert the relevant columns to a numeric type by removing the thousands separators and then impute the missing values with the mean.



```python
columns_to_convert = [
    'persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap',
    'hpp_rawat_jalan', 'total_hpp', 'pendapatan_operasional',
    'pendapatan_penjualan', 'ppn_keluaran', 'ppn_masukan', 'ppn_netto',
    'laba_kotor_penjualan', 'beban_operasional'
]

for col in columns_to_convert:
    df_cleaned[col] = df_cleaned[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    df_cleaned[col] = pd.to_numeric(df_cleaned[col])

for col in columns_to_convert:
     df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mean())

print("\nMissing values after conversion and imputation:")
print(df_cleaned.isnull().sum())

display(df_cleaned.head())
```

**Reasoning**:
The conversion to numeric failed because of the 'nan' strings resulting from missing values. Handle the missing values first by filling them with a placeholder string before attempting the conversion, and then convert to numeric and impute the mean for the remaining missing values.



```python
columns_to_convert = [
    'persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap',
    'hpp_rawat_jalan', 'total_hpp', 'pendapatan_operasional',
    'pendapatan_penjualan', 'ppn_keluaran', 'ppn_masukan', 'ppn_netto',
    'laba_kotor_penjualan', 'beban_operasional'
]

for col in columns_to_convert:
    df_cleaned[col] = df_cleaned[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

for col in columns_to_convert:
     df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mean())

print("\nMissing values after conversion and imputation:")
print(df_cleaned.isnull().sum())

display(df_cleaned.head())
```

**Reasoning**:
The 'tahun' column still contains missing values. Since 'tahun' represents a year, imputing with the mean might not be the most appropriate strategy. Given the small number of missing values (6), dropping the rows with missing 'tahun' values is a reasonable approach.



```python
df_cleaned.dropna(subset=['tahun'], inplace=True)

print("\nMissing values after dropping rows with missing 'tahun':")
print(df_cleaned.isnull().sum())

display(df_cleaned.head())
```

## Data analysis

### Subtask:
Perform data analysis on the cleaned data to gain insights into the financial data over the years.


**Reasoning**:
Calculate the required aggregations for each year and store them in new data structures.



```python
yearly_total_hpp = df_cleaned.groupby('tahun')['total_hpp'].sum()
yearly_pendapatan_operasional = df_cleaned.groupby('tahun')['pendapatan_operasional'].sum()
yearly_beban_operasional = df_cleaned.groupby('tahun')['beban_operasional'].sum()
yearly_average_laba_kotor_penjualan = df_cleaned.groupby('tahun')['laba_kotor_penjualan'].mean()
yearly_total_ppn_netto = df_cleaned.groupby('tahun')['ppn_netto'].sum()

print("Yearly Total HPP:")
display(yearly_total_hpp)
print("\nYearly Pendapatan Operasional:")
display(yearly_pendapatan_operasional)
print("\nYearly Beban Operasional:")
display(yearly_beban_operasional)
print("\nYearly Average Laba Kotor Penjualan:")
display(yearly_average_laba_kotor_penjualan)
print("\nYearly Total PPN Netto:")
display(yearly_total_ppn_netto)
```

## Data visualization

### Subtask:
Create colorful visualizations (graphs) to represent the findings from the data analysis.


**Reasoning**:
Generate colorful visualizations for the yearly financial metrics using bar plots as instructed.



```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(10, 25))

sns.barplot(x=yearly_total_hpp.index, y=yearly_total_hpp.values, ax=axes[0], palette='viridis')
axes[0].set_title('Total HPP per Tahun')
axes[0].set_xlabel('Tahun')
axes[0].set_ylabel('Total HPP')

sns.barplot(x=yearly_pendapatan_operasional.index, y=yearly_pendapatan_operasional.values, ax=axes[1], palette='viridis')
axes[1].set_title('Pendapatan Operasional per Tahun')
axes[1].set_xlabel('Tahun')
axes[1].set_ylabel('Pendapatan Operasional')

sns.barplot(x=yearly_beban_operasional.index, y=yearly_beban_operasional.values, ax=axes[2], palette='viridis')
axes[2].set_title('Beban Operasional per Tahun')
axes[2].set_xlabel('Tahun')
axes[2].set_ylabel('Beban Operasional')

sns.barplot(x=yearly_average_laba_kotor_penjualan.index, y=yearly_average_laba_kotor_penjualan.values, ax=axes[3], palette='viridis')
axes[3].set_title('Rata-rata Laba Kotor Penjualan per Tahun')
axes[3].set_xlabel('Tahun')
axes[3].set_ylabel('Rata-rata Laba Kotor Penjualan')

sns.barplot(x=yearly_total_ppn_netto.index, y=yearly_total_ppn_netto.values, ax=axes[4], palette='viridis')
axes[4].set_title('Total PPN Netto per Tahun')
axes[4].set_xlabel('Tahun')
axes[4].set_ylabel('Total PPN Netto')

plt.tight_layout()
plt.show()
```
