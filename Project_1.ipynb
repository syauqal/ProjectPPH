## Data cleaning

### Subtask:
Clean the data by handling missing values and inconsistencies. Given the significant number of missing values identified in the exploration phase, focus on strategies like imputation or removal of columns/rows if necessary, based on the nature and extent of missing data in each column.


**Reasoning**:
Based on the missing value count from the previous step, several columns have a significant number of missing values. I will analyze the percentage of missing values per column and decide on a cleaning strategy (dropping or imputing) based on the percentage and the nature of the column.



```python
missing_percentage = df.isnull().sum() / len(df) * 100
print("Percentage of missing values per column:")
print(missing_percentage)

columns_to_drop = missing_percentage[missing_percentage > 50].index.tolist()
print(f"\nColumns to drop (more than 50% missing): {columns_to_drop}")

df_cleaned = df.drop(columns=columns_to_drop)

# Impute missing values in remaining columns.
# For numerical columns, use the median as it's less sensitive to outliers.
# For object columns, use the mode.
for column in df_cleaned.columns:
    if df_cleaned[column].isnull().sum() > 0:
        if df_cleaned[column].dtype in ['float64', 'int64']:
            median_val = df_cleaned[column].median()
            df_cleaned[column].fillna(median_val, inplace=True)
            print(f"Imputed missing values in '{column}' with median ({median_val}).")
        elif df_cleaned[column].dtype == 'object':
            mode_val = df_cleaned[column].mode()[0]
            df_cleaned[column].fillna(mode_val, inplace=True)
            print(f"Imputed missing values in '{column}' with mode ({mode_val}).")

print("\nMissing values after cleaning:")
print(df_cleaned.isnull().sum())

display(df_cleaned.head())
```

**Reasoning**:
The previous step successfully dropped columns with more than 50% missing values and imputed missing values in the remaining columns. The output confirms that there are no more missing values in the `df_cleaned` DataFrame. The next step is to confirm that the data types are correct and handle any remaining inconsistencies as per the instructions. The `df.info()` and `df.describe()` outputs from the exploration phase will be used to guide this process.



```python
print("Data types after cleaning:")
df_cleaned.info()

print("\nDescriptive statistics after cleaning:")
display(df_cleaned.describe())

# Check for inconsistencies in 'ppn_netto' which was identified as 'object' type.
# Based on the head, it contains numerical values and '-   '.
# Convert 'ppn_netto' to numeric, coercing errors to NaN, and then impute NaNs.
df_cleaned['ppn_netto'] = pd.to_numeric(df_cleaned['ppn_netto'], errors='coerce')

# Impute remaining NaNs in 'ppn_netto' with the median
if df_cleaned['ppn_netto'].isnull().sum() > 0:
    median_ppn_netto = df_cleaned['ppn_netto'].median()
    df_cleaned['ppn_netto'].fillna(median_ppn_netto, inplace=True)
    print(f"\nImputed remaining missing values in 'ppn_netto' with median ({median_ppn_netto}).")

print("\nData types after converting 'ppn_netto':")
df_cleaned.info()

print("\nMissing values after converting 'ppn_netto':")
print(df_cleaned.isnull().sum())

display(df_cleaned.head())
```

**Reasoning**:
The previous steps handled missing values and converted the 'ppn_netto' column to numeric. However, the `df_cleaned.info()` output shows that most columns are still of 'object' type despite containing numerical values. This indicates further data type inconsistencies that need to be addressed. I will iterate through the remaining 'object' type columns and attempt to convert them to a suitable numeric type, coercing errors to NaN and then imputing any resulting NaNs.



```python
# Identify columns that are still 'object' type but should be numeric
object_cols_to_convert = df_cleaned.select_dtypes(include='object').columns

print(f"\nAttempting to convert the following object columns to numeric: {list(object_cols_to_convert)}")

for col in object_cols_to_convert:
    # Attempt to convert to numeric, coercing errors to NaN
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

    # Impute any NaNs created by the coercion with the median
    if df_cleaned[col].isnull().sum() > 0:
        median_val = df_cleaned[col].median()
        df_cleaned[col].fillna(median_val, inplace=True)
        print(f"Imputed missing values in '{col}' with median ({median_val}) after numeric conversion.")

print("\nData types after attempting numeric conversion:")
df_cleaned.info()

print("\nMissing values after attempting numeric conversion:")
print(df_cleaned.isnull().sum())

display(df_cleaned.head())
```

## Data analysis

### Subtask:
Perform comprehensive data analysis on the cleaned data, including calculating key metrics, identifying trends, and exploring relationships relevant to accounting and tax.


**Reasoning**:
Calculate the sum of key financial metrics over the years and analyze their trends. The key financial metrics based on the available columns are total HPP, total operating revenue (sum of pendapatan_operasional and pendapatan_penjualan), total beban_operasional, and total PPN_netto.



```python
# Calculate total operating revenue
df_cleaned['total_operating_revenue'] = df_cleaned['pendapatan_operasional'] + df_cleaned['pendapatan_penjualan']

# Group by year and sum the key financial metrics
annual_financial_summary = df_cleaned.groupby('tahun').agg({
    'total_hpp': 'sum',
    'total_operating_revenue': 'sum',
    'beban_operasional': 'sum',
    'ppn_netto': 'sum'
}).reset_index()

print("Annual Financial Summary:")
display(annual_financial_summary)

# Analyze trends (simple percentage change year-over-year)
annual_financial_trends = annual_financial_summary.set_index('tahun').pct_change() * 100
print("\nAnnual Financial Trends (Year-over-Year % Change):")
display(annual_financial_trends)
```

**Reasoning**:
Calculate profitability ratios (gross profit margin, operating profit margin) and analyze their trends. Gross profit can be calculated as total_operating_revenue - total_hpp. Operating profit can be calculated as gross_profit - beban_operasional.



```python
# Calculate Gross Profit and Operating Profit
annual_financial_summary['gross_profit'] = annual_financial_summary['total_operating_revenue'] - annual_financial_summary['total_hpp']
annual_financial_summary['operating_profit'] = annual_financial_summary['gross_profit'] - annual_financial_summary['beban_operasional']

# Calculate Profitability Ratios
# Gross Profit Margin = (Gross Profit / Total Operating Revenue) * 100
annual_financial_summary['gross_profit_margin'] = (annual_financial_summary['gross_profit'] / annual_financial_summary['total_operating_revenue']) * 100

# Operating Profit Margin = (Operating Profit / Total Operating Revenue) * 100
annual_financial_summary['operating_profit_margin'] = (annual_financial_summary['operating_profit'] / annual_financial_summary['total_operating_revenue']) * 100

print("\nAnnual Financial Summary with Profitability Ratios:")
display(annual_financial_summary)

# Analyze trends in profitability ratios (simple percentage change year-over-year)
# It's more meaningful to see the trend of the ratio itself, not the percentage change of the ratio
# So, we will just display the ratios over the years
print("\nAnnual Profitability Ratios Trends:")
display(annual_financial_summary[['tahun', 'gross_profit_margin', 'operating_profit_margin']])
```

**Reasoning**:
Explore the relationship between different financial components, specifically the impact of operating expenses on profit before tax. Although 'laba_sebelum_pajak' was dropped due to missing values, operating profit is a close proxy.
```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Operating Profit
ax1.plot(annual_financial_summary['tahun'], annual_financial_summary['operating_profit'], marker='o', linestyle='-', color='tab:blue', label='Operating Profit')
ax1.set_xlabel('Year')
ax1.set_ylabel('Operating Profit', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.set_title('Operating Profit and Operating Expenses Over Years')
ax1.legend(loc='upper left')

# Create a second y-axis for Operating Expenses
ax2 = ax1.twinx()
ax2.plot(annual_financial_summary['tahun'], annual_financial_summary['beban_operasional'], marker='x', linestyle='--', color='tab:red', label='Operating Expenses')
ax2.set_ylabel('Operating Expenses', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')
ax2.legend(loc='upper right')

# Improve x-axis ticks
plt.xticks(annual_financial_summary['tahun'])

plt.tight_layout()
plt.show()
```
